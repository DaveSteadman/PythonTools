```python
# Create a multi-scale icon file, from an input PNG file.

from PIL import Image
import os

# Input PNG file path.
input_png_path = "Icon_x1000.png"
# Output ICO file path.
output_ico_path = "Icon.ico"


# open the input PNG file
with Image.open(input_png_path) as img:

    # Define the sizes for the ICO file.
    sizes = [(16, 16), (32, 32), (48, 48), (64, 64), (128, 128), (256, 256)]

    # Save the image as an ICO file with multiple sizes.
    img.save(output_ico_path, format='ICO', sizes=sizes)
    print(f"Created ICO file: {output_ico_path} with sizes: {sizes}")

```
